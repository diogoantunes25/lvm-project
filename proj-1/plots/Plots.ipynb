```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("../data/data.csv")
df["size"] = df["width"] + df["height"]
df["sating"] = (df["solving"] / (df["solving"] + df["init"])) * 100
df
```

```python
plt.rcParams["figure.figsize"] = (15,4)

plt.subplot(1, 2, 1)
sns.barplot(df, x = "size", y = "init", hue = "Algorithm")
plt.xlabel("Puzzle size")
plt.ylabel("Initialization time (sec)")

plt.subplot(1, 2, 2)
sns.barplot(df, x = "size", y = "solving", hue = "Algorithm")
plt.xlabel("Puzzle size")
plt.ylabel("Solving time (sec)")
```

```python
plt.rcParams["figure.figsize"] = (7,4)

sns.barplot(df, x = "size", y = "sating", hue = "Algorithm")
plt.xlabel("Puzzle size")
plt.ylabel("Percentage of time in SAT solver")
```
